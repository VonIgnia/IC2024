```python
#imports
from pyedflib import highlevel
import pyedflib as plib
import numpy as np
import matplotlib.pyplot as plt
#from scipy.fft import fft, ifft,rfft
```

```python
path = "1.Dados_Anteriores\Raw Empirical EEG Data\EEG_Cat_Study4_Resting_S1.bdf" 
signals, signal_headers, header = highlevel.read_edf(path)

# Sample data setup (replace this with your actual signals and signal_headers)
#fs = 256 # Sampling frequency (Hz)
fs = 2048
T = 1.0 / fs  # Sampling period


# Example signal data and headers

# Define FFT function
def compute_fft(signal, fs):
    n = len(signal)
    fft_result = np.fft.fft(signal)
    fft_freq = np.fft.fftfreq(n, T)
    return fft_freq, np.abs(fft_result) # Only take positive frequencies
    #return fft_freq[:n // 2], np.abs(fft_result)[:n // 2]  # Only take positive frequencies

# Plotting time-domain and frequency-domain (FFT) analysis
fig, (ax_time, ax_freq) = plt.subplots(2, 1, figsize=(12, 8))

for i in range(len(signals)):
    if signal_headers[i]["label"] == "C1": 
        n_samples = len(signal_headers[i])  # Number of samples
        label = signal_headers[i]["label"]
        #print(label)

        # Time-domain plot
        ax_time.plot(signals[i], label=label)
        
        # Frequency-domain plot
        fft_freq, fft_magnitude = compute_fft(signals[i], fs)
        ax_freq.plot(fft_freq, fft_magnitude, label=f'{label} (FFT)')

# Time-domain plot settings
ax_time.set_title('Time Domain Signals')
ax_time.set_xlabel('Time (seconds)')
ax_time.set_ylabel('Amplitude')
ax_time.legend()
ax_time.grid()

# Frequency-domain plot settings
ax_freq.set_title('Frequency Domain (FFT) Analysis')
ax_freq.set_xlabel('Frequency (Hz)')
ax_freq.set_ylabel('Magnitude')
ax_freq.set_xlim(-1,20)
ax_freq.legend()
ax_freq.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from ipywidgets import interact 
%matplotlib inline 
  
x = np.linspace(0, 5, 100) 
y = np.exp(x) 
  
# Plotting the graph for exponential function 
def plot(checkbox): 
      
    # if checkbox is ticked then scatter 
    # plot will be displayed 
    if checkbox: 
        plt.scatter(x, y, s = 5) 
        plt.title('Scatter plot of exponential curve') 
      
    # if checkbox is not ticked (by default) 
    # line plot will be displayed 
    else: 
        plt.plot(x, y) 
        plt.title('Line plot of exponential curve') 
          
# calling the interact function         
interact(plot, checkbox = bool()) 
```
